```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import requests
from io import StringIO
```

```python
url = "http://socr.ucla.edu/docs/resources/SOCR_Data/SOCR_Data_Dinov_020108_HeightsWeights.html"
response = requests.get(url)
html = response.text
dfs = pd.read_html(html)
df = dfs[0]
df.columns = ['Index', 'Height', 'Weight']
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y='Height', data=df)
plt.title('Box Plot for Height')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(y='Height', data=df)
plt.title('Violin Plot for Height')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from IPython.display import display

url = "https://en.wikipedia.org/wiki/List_of_highest-grossing_films_in_China"
df = pd.read_html(url, header=0)[0]

numerical_columns = df.select_dtypes(include=['number']).columns
result_table = pd.DataFrame(columns=['No', 'Название переменной', 'визуальной проверки нормальности данных', 'сюжет QQ', 'Тест Шапиро-Вилка', "K'2-тест Д'Агостино", 'Тест Андерсона-Дарлинга'])

for i, col in enumerate(numerical_columns):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.hist(df[col], bins=20, edgecolor='black')
    plt.title(f'Histogram of {col}')

    plt.subplot(1, 2, 2)
    stats.probplot(df[col], dist="norm", plot=plt)
    plt.title(f'QQ Plot of {col}')

    plt.show()

    result_table = result_table.append({
        'No': i + 1,
        'Название переменной': col,
        'визуальной проверки нормальности данных': 'гистограмма и QQ-plot',
        'сюжет QQ': 'гистограмма и QQ-plot',
        'Тест Шапиро-Вилка': stats.shapiro(df[col])[1],
        "K'2-тест Д'Агостино": stats.normaltest(df[col])[1],
        'Тест Андерсона-Дарлинга': stats.anderson(df[col])[1]
    }, ignore_index=True)

display(result_table)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

file_path = "/content/Variant_3.csv"
df = pd.read_csv(file_path, delimiter=';', skiprows=15)

df.columns = ["Date", "Max Temperature", "Min Temperature", "Avg Temperature",
              "Atmospheric Pressure", "Wind Speed", "Precipitation", "Effective Temperature"]

for col in df.columns[1:]:
    if df[col].dtype == object:
        df[col] = df[col].str.replace(',', '.').astype(float)

df['Date'] = pd.to_datetime(df['Date'], format='%d.%m.%Y')
df_algeria = df[(df['Date'] >= '2010-12-01') & (df['Date'] <= '2020-12-01')]

columns_of_interest = ["Max Temperature", "Min Temperature", "Avg Temperature",
                         "Atmospheric Pressure", "Wind Speed", "Precipitation", "Effective Temperature"]
data_to_normalize = df_algeria[columns_of_interest]
data_to_normalize = data_to_normalize.fillna(0)

scaler_minmax = MinMaxScaler()
normalized_data_min = scaler_minmax.fit_transform(data_to_normalize)

scaler_standard = StandardScaler()
standardized_data = scaler_standard.fit_transform(data_to_normalize)

results_table = pd.DataFrame({
    'нормализация min': normalized_data_min.min(axis=0),
    'нормализация max': normalized_data_min.max(axis=0),
    'Стандартизация среднее': standardized_data.mean(axis=0),
    'Стандартизация Стандартное отклонение': standardized_data.std(axis=0)
}, index=columns_of_interest)

from IPython.display import display
pd.set_option('display.float_format', '{:.4f}'.format)
display(results_table)
```
